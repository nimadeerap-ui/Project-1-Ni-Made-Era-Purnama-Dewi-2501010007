Hitung Jumlah Mahasiswa Lulus

```python
import numpy as np
nilai = np.array([88,72,65,91,58,84,77,89,45,95])
lulus = nilai[nilai >=60]
jumlah_lulus = len(lulus)
jumlah_tidak_lulus = len(nilai) - jumlah_lulus

print(f"Mahasiswa Lulus: {jumlah_lulus}")
print(f"Mahasiswa Tidak lulus: {jumlah_tidak_lulus}")
```

Grafik Tertinggi dan Terendah

```python
import numpy as np
import matplotlib.pyplot as plt

nilai = np.array([88,72,65,91,58,84,77,89,45,95])

n_max = np.max(nilai)
n_min = np.min(nilai)

plt.figure(figsize=(6, 4))
plt.bar(['Tertinggi', 'Terendah'], [n_max, n_min], color=['green', 'red'])
plt.title('Perbandingan Nilai Tertinggi dan Terendah')
plt.ylabel('Nilai')
plt.show()
```

Grafik Data Kelulusan

```python
labels = ['Lulus (>=60)', 'Tidak Lulus (<=60)']
sizes = [jumlah_lulus, jumlah_tidak_lulus]
colors = ['blue', 'yellow']

plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=40, colors=colors)
plt.title('Persentase Kelulusan Mahasiswa')
plt.show()
```
